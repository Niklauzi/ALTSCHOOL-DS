## Communicating with Data

Intuition: Clear communication through graphical means

```python
# Sale/Marketing Analyst.
```

```python
# import packages
import matplotlib.pyplot as plt # static & interactive plotting
import pandas as pd # Data manipulation 
import numpy as np # Numerical analysis
```

```python
# Sample data 
# Creating a dataframe using a dictionary 
data = pd.DataFrame({'Year': [2017, 2018 ,2019 ,2020 ,2021], 
                    'Sales': [350, 400, 450, 500, 600]})
```

```python
# Line plot works effectively for time, trends..
plt.figure(figsize=(10,6)) # specifying your plot/figure size
plt.plot(data['Year'], data['Sales'], marker = 'o') # Plot Year & Sales, markers
plt.title('Annual Sales Trend') # Add a title
plt.xlabel('Year') # Label the x axis
plt.ylabel('Sales (in thousands)') # Label the y axis
plt.grid(True) # Add a grid
plt.show() # Display the plot
```

## Understanding Context(Exploratory & Explanatory)

Exploratory: Uncover patterns, relationships, outliers/anomalies 

Explanatory: Presenting findings, showing insights  

```python
# import seaborn 
import seaborn as sns # package for statistical viz

# Sample data 
tips = sns.load_dataset('tips') # seaborn's data repository 
```

```python
tips.head() # reads the first five rows of the dataframe
```

```python
# Exploratory 
# Pairplot
plt.figure(figsize=(10,6)) # creating a figure size
sns.pairplot(tips) # pairplot function/method in seaborn/pairwise relationship
plt.title('Exploratory Data Analysis of Restaurant Data') # Graph title
plt.show() # Display the plot
```

```python
# Explanatory
# barplots
plt.figure(figsize=(10, 6)) # Figure size(10 inches by 6 inches)
sns.barplot(x='day', y='total_bill', data=tips, estimator=sum, palette='viridis') 
# reps the sum of the 'total_bill'/ color palette
plt.title('Total Bill per Day') # title of the chart
plt.xlabel('Day of the Week') # x axis label
plt.ylabel('Total Bill (Sum)') # y axis label
plt.show() # Display the plot
```

## Eliminating clutter 

Intuition: Reduce cognitive brainpower for learning new Information, do not confuse your viewers, less distraction 
    
Encourage: Simplicity, focus, minimalism, consistency 

```python
# Sample data
# Created a dataframe from a dictionary 
data = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'Value': [10, 20, 30, 40, 50]
})
```

```python
# Cluttered bar plot
plt.figure(figsize=(10, 6)) # figure size(10,6)
bars = plt.bar(data['Category'], data['Value'], color='skyblue') # category, values, color = skyblue

# Adding unnecessary elements
plt.title('Bar Plot with Clutter', fontsize=20, fontweight='bold') # fontsize, fontweight
plt.xlabel('Category', fontsize=15) # x axis label font size
plt.ylabel('Value', fontsize=15) # y axis label font size
plt.grid(True, which='both', linestyle='--', linewidth=0.5) # major/minor ticks
plt.xticks(fontsize=12, rotation=45) # fontsize & rotate our label
plt.yticks(fontsize=12) # font size
plt.legend(['Values'], fontsize=12) # font size
plt.axhline(25, color='red', linewidth=1.5, linestyle='--') # horizontal axes line
plt.axhline(45, color='green', linewidth=1.5, linestyle='--') # horizontal axes lines
plt.text(-0.2, 55, 'Extra Text', fontsize=12) # Adding unnecessary text

plt.show() # Display of chart
```

```python
# Code for Cleaned Visualization:
plt.figure(figsize=(10, 6)) # figure size(10 inches by 6 inches)
bars = plt.bar(data['Category'], data['Value'], color='skyblue') # barplot for category and value

# Simplifying the plot
plt.title('Cleaned Bar Plot', fontsize=16) # title, font size
plt.xlabel('Category', fontsize=14) # x axis label
plt.ylabel('Value', fontsize=14) # y axis label
plt.xticks(fontsize=10) # set font size for x ticks
plt.yticks(fontsize=10) # set font size for y ticks

# Removing unnecessary elements
plt.grid(False) # remove grids
plt.legend().remove() # remove legends
plt.axhline(25, color='red', linewidth=1, linestyle='--') # Horizontal lines
plt.axhline(45, color='green', linewidth=1, linestyle='--') # Horizontal lines

plt.show() # Display plot
```

## Focusing Attention 

Intuition: Direct your viewers attention, to key messages, convey important insights
    
    
To achieve this: Highlighting, Annotations, contrasting colors, visual hierarchy 

```python
# Sample data
tips = sns.load_dataset('tips') # tips data from seaborn repository 

# Highlighting a specific data point
plt.figure(figsize=(10, 6)) # figure size (10, 6 inches)
sns.barplot(x='day', y='total_bill', data=tips, estimator=sum, palette='muted') # barplot

# Highlight Thursday
highlight = tips[tips['day'] == 'Thur']['total_bill'].sum() # filtering the the day column 
plt.bar('Thur', highlight, color='red') # Highlighting with the colour red

plt.title('Total Bill per Day') # title of the graphs
plt.xlabel('Day of the Week') # x axis label
plt.ylabel('Total Bill (Sum)') # y xis label
plt.show()
```

```python
# Illustrating annotations 
```

```python
# Generate sample data
np.random.seed(0) # ensures reproducibility 
x = np.random.rand(10) # random number Generator(10, between 0,1), x axis
y = np.random.rand(10) # random number Generator(10, between 0,1), y axis
sizes = 1000 * np.random.rand(10)  # Bubble sizes # random number Generator(10, between 0,1)

# Create a scatter plot
plt.figure(figsize=(10, 6)) # figure size
plt.scatter(x, y, s=sizes, alpha=0.5, c='blue') # x,y Coordinate, size, color
plt.title('Scatter Plot with Annotations') # title
plt.xlabel('X-axis') # x axis label
plt.ylabel('Y-axis') # y axis label

# Adding annotations
for i, txt in enumerate(range(10)): # for loop to generate 0-9, enumeration
    plt.annotate(f'Point {txt}', (x[i], y[i]), # create points as annotations 
                 textcoords="offset points",
                 xytext=(0, 10),
                 ha='center',
                 fontsize=9,
                 color='red') # Specifies that the annotation's position is offset in points
# Specifies the offset (0 points horizontally and 10 points vertically) for the annotation text.
# Horizontally align the annotation text to the center of the point.
# Sets the font size of the annotation text.
# Sets the color of the annotation text to red.

# Highlight a specific point
highlight_idx = 5 # Specifies the index of the point to be highlighted.
plt.scatter(x[highlight_idx], 
            y[highlight_idx],
            s=sizes[highlight_idx],
            color='red') # Highlight the specific point by plotting it again in red.
# Scatter plots for annotations
plt.annotate('Highlighted Point', (x[highlight_idx], y[highlight_idx]),
             textcoords="offset points",
             xytext=(0, 15),
             ha='center',
             fontsize=12,
             color='red',
             weight='bold') # Add a prominent annotation to the highlighted point

plt.grid(True) # Grid
plt.show() # Display the chart
```

## Telling a story

Intuition: Tell a compelling story, lead your viewers, create compelling narratives 
    
    
To achieve this: Context, flow, Emotions, Engagement 

```python
# Sample data
data = pd.DataFrame({
    'Year': [2017, 2018, 2019, 2020, 2021],
    'Sales': [350, 400, 450, 500, 600]
}) # creating a data frame using a dictionary 

# Storytelling with a line plot
plt.figure(figsize=(10, 6)) # figure size (10, 6 inches)
plt.plot(data['Year'], data['Sales'], marker='o', color='b') # Creates a line plot with 'Year' on the x-axis and 'Sales' on the y-axis.
plt.title('Annual Sales Trend', fontsize=16) # Adds a title to the plot with a font size of 16.
plt.xlabel('Year', fontsize=14) # Labels the x-axis as "Year" with a font size of 14.
plt.ylabel('Sales (in thousands)', fontsize=14) # Labels the y-axis as "Sales (in thousands)" with a font size of 14.

# Adding annotations to tell a story
for i, txt in enumerate(data['Sales']):
    plt.annotate(txt, (data['Year'][i], data['Sales'][i]),
                 textcoords="offset points",
                 xytext=(0, 10),
                 ha='center') # Iterates over the 'Sales' column, where i is the index and txt is the sales value.
# Add annotations to each data point in the plot.

plt.grid(True) # Adds grid lines to the plot for better readability.
plt.show() # Display the plot
```
