## Gestalt Principles of Data Visualisation 

```python
# The Gestalt principles helping understand how humans perceive visual elements
# This is to make our visuals effective and understandable 
```

### 1. Principle of Proximity 

```python
# Objects close to each other are seen as related or part of the same group
```

```python
# import packages
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# Sample data
data = pd.DataFrame({
    'Category': ['A']*50 + ['B']*50,
    'Value1': np.random.randn(100),
    'Value2': np.random.randn(100)
})
```

```python
data.head()
```

```python
# Scatter plot with proximity principle
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Value1', y='Value2', hue='Category', data=data)
plt.title('Principle of Proximity')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.rand(10,2)
cluster1 = data + np.array([-0.5, 0.5])
cluster2 = data + np.array([1, -1])

plt.scatter(cluster1[:, 0], cluster1[:, 1], color='blue', label='Group 1')
plt.scatter(cluster2[:, 0], cluster2[:, 1], color='red', label='Group 2')
plt.legend()
plt.title('Gestalt Principle of Proximity')
plt.show()
```

### 2. Principle of Similarity

```python
# Objects that look similar are perceived as the same group or part of the same group
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Value1', y='Value2', hue='Category', style='Category', data=data)
plt.title('Principle of Similarity')
plt.show()
```

### 3. Principle of Enclosure

```python
# Objects that are enclosed by a border are perceived as a group
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Value1', y='Value2', hue='Category', data=data)
plt.title('Principle of Enclosure')

# Adding enclosure
plt.gca().add_patch(plt.Rectangle((0, 0), 2, 2, fill=False, edgecolor='red', linewidth=2))
plt.show()
```

### 4. Principle of Continuity 

```python
# The human eyes naturally follow lines and curves. When sharing continuous values, use charts similar to lines
# and curves 
```

```python
# Sample data for line plot
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Line plot with continuity principle
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label='Sine Wave')
plt.plot(x, y2, label='Cosine Wave')
plt.title('Principle of Continuity')
plt.legend()
plt.show()
```

### 5. Principle of Closure

```python
# Our brain tends to fill in the gaps in incomplete shapes or lines 
# The brain will complete the object
# Line plot with closure principle
plt.figure(figsize=(10, 6))
plt.plot([0, 2, 2, 0, 0], [0, 0, 2, 2, 0], linestyle='--', color='blue')
plt.fill([0, 2, 2, 0, 0], [0, 0, 2, 2, 0], alpha=0.3)
plt.title('Principle of Closure')
plt.show()
```

### 6. Principle of Connection 

```python
# Objects that are connected by lines are seen as related 
np.random.seed(42)
data = pd.DataFrame({
    'Group': ['A', 'A', 'B', 'B'],
    'X': np.random.rand(4),
    'Y': np.random.rand(4)
})

# Scatter plot with connection principle
plt.figure(figsize=(10, 6))
sns.scatterplot(x='X', y='Y', hue='Group', style='Group', data=data, s=100)
for name, group in data.groupby('Group'):
    plt.plot(group['X'], group['Y'], label=name)
plt.title('Principle of Connection')
plt.legend()
plt.show()
```
