```python
# import packages 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
# Anscombe's Quartet
# Create the data for Anscombe's Quartet
data = {
    'I': {
        'x': [10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5],
        'y': [8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68]
    },
    'II': {
        'x': [10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5],
        'y': [9.14, 8.14, 8.74, 8.77, 9.26, 8.10, 6.13, 3.10, 9.13, 7.26, 4.74]
    },
    'III': {
        'x': [10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5],
        'y': [7.46, 6.77, 12.74, 7.11, 7.81, 8.84, 6.08, 5.39, 8.15, 6.42, 5.73]
    },
    'IV': {
        'x': [8, 8, 8, 8, 8, 8, 8, 8, 8, 8, 8],
        'y': [6.58, 5.76, 7.71, 8.84, 8.47, 7.04, 5.25, 5.56, 7.91, 6.89, 7.92]
    }
}
```

```python
data 
```

```python
# Statistical summaries
stats = {}

for key in data.keys():
    x = np.array(data[key]['x'])
    y = np.array(data[key]['y'])
    df = pd.DataFrame({'x': x, 'y': y})
    model = ols("y ~ x", df).fit()
    stats[key] = {
        'mean_x': np.mean(x),
        'mean_y': np.mean(y),
        'variance_x': np.var(x, ddof=1),
        'variance_y': np.var(y, ddof=1),
        'correlation': np.corrcoef(x, y)[0, 1],
        'regression_slope': model.params['x'],
        'regression_intercept': model.params['Intercept']
    }
```

```python
stats
```

```python
stats_df = pd.DataFrame(stats).T
```

```python
stats_df
```

```python
# Creating a figure
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes = axes.flatten()

for i, key in enumerate(data.keys()):
    x = np.array(data[key]['x'])
    y = np.array(data[key]['y'])
    axes[i].scatter(x, y)
    axes[i].set_xlim(0, 20)
    axes[i].set_ylim(0, 15)
    axes[i].set_title(f'Dataset {key}')
    # Plotting the linear regression line
    sns.regplot(x=x, y=y, ax=axes[i], ci=None, line_kws={"color":"orange"})

plt.tight_layout()
plt.show()
```

```python
# Pareto plot/chart
df = pd.DataFrame({'count': [100, 56, 67, 89, 90, 67, 24]})
```

```python
df.index=['B', 'C', 'D', 'A', 'E', 'F', 'G']
```

```python
df 
```

```python
#sort DataFrame by count descending
df = df.sort_values(by='count', ascending=False)

#add column to display cumulative percentage
df['cumperc'] = df['count'].cumsum()/df['count'].sum()*100

#view DataFrame
df
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter

#define aesthetics for plot
color1 = 'steelblue'
color2 = 'red'
line_size = 4

#create basic bar plot
fig, ax = plt.subplots()
ax.bar(df.index, df['count'], color=color1)

#add cumulative percentage line to plot
ax2 = ax.twinx()
ax2.plot(df.index, df['cumperc'], color=color2, marker="D", ms=line_size)
ax2.yaxis.set_major_formatter(PercentFormatter())

#specify axis colors
ax.tick_params(axis='y', colors=color1)
ax2.tick_params(axis='y', colors=color2)

#display Pareto chart
plt.show()
```
