## Communicating with Data

Intuition: Clear communication through graphical means

```python
# Sale/Marketing Analyst.
```

```python
# import packages
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
```

```python
# Sample data 
data = pd.DataFrame({'Year': [2017, 2018 ,2019 ,2020 ,2021],
                    'Sales': [350, 400, 450, 500, 600]})
```

```python
data
```

```python
# Line plot works effectively for time, trends..
plt.figure(figsize=(10,6))
plt.plot(data['Year'], data['Sales'], marker = 'o')
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: '{:.0f}'.format(x)))

#data.plot(x='Year', y='Sales', marker = 'o')

plt.title('Annual Sales Trend')
plt.xlabel('Year')
plt.ylabel('Sales (in thousands)')
plt.grid(True)
plt.show()
```

## Understanding Context(Exploratory & Explanatory)

Exploratory: Uncover patterns, relationships, outliers/anomalies 

Explanatory: Presenting findings, showing insights  

```python
# import seaborn 
import seaborn as sns 

# Sample data 
tips = sns.load_dataset('tips')
```

```python
tips.head()  
```

```python
# Exploratory 
# Pairplot
plt.figure(figsize=(10,6))
sns.pairplot(tips)
plt.title('Exploratory Data Analysis of Restaurant Data')
plt.show()
```

```python
# Explanatory
# barplots
plt.figure(figsize=(10, 6))
sns.barplot(x='day', y='total_bill', data=tips, estimator=sum, palette='viridis')
plt.title('Total Bill per Day')
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill (Sum)')
plt.show()
```

## Choosing Approprite Visual Display

Intuition: The right type of visuals, dependent on data types, the message being passed
    
Data types: categorical - barplots, continuous data type - scatterplots
    
Message: Comparison - barplot, Distribution - histograms, trends - lineplots

```python
# Message is comparison 
# Data type is categories
# Intuition - barplot
# Bar plot example
plt.figure(figsize=(10, 6))
sns.barplot(x='day', y='total_bill', data=tips, estimator=sum, palette='muted')
plt.title('Total Bill per Day')
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill (Sum)')
plt.show()
```

## Eliminating clutter 

Intuition: Reduce cognitive brainpower for learning new Information, do not confuse your viewers, less distraction 
    
Encourage: Simplicity, focus, minimalism, consistency 

```python
# Sample data
data = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'Value': [10, 20, 30, 40, 50]
})
```

```python
# Cluttered bar plot
plt.figure(figsize=(10, 6))
bars = plt.bar(data['Category'], data['Value'], color='skyblue')

# Adding unnecessary elements
plt.title('Bar Plot with Clutter', fontsize=20, fontweight='bold')
plt.xlabel('Category', fontsize=15)
plt.ylabel('Value', fontsize=15)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.xticks(fontsize=12, rotation=45)
plt.yticks(fontsize=12)
plt.legend(['Values'], fontsize=12)
plt.axhline(25, color='red', linewidth=1.5, linestyle='--')
plt.axhline(45, color='green', linewidth=1.5, linestyle='--')
plt.text(-0.2, 55, 'Extra Text', fontsize=12)

plt.show()
```

```python
# Code for Cleaned Visualization:
plt.figure(figsize=(10, 6))
bars = plt.bar(data['Category'], data['Value'], color='skyblue')

# Simplifying the plot
plt.title('Cleaned Bar Plot', fontsize=16)
plt.xlabel('Category', fontsize=14)
plt.ylabel('Value', fontsize=14)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# Removing unnecessary elements
plt.grid(False)
plt.legend().remove()
plt.axhline(25, color='red', linewidth=1, linestyle='--')
plt.axhline(45, color='green', linewidth=1, linestyle='--')

plt.show()
```

## Focusing Attention 

Intuition: Direct your viewers attention, to key messages, convey important insights
    
    
To achieve this: Highlighting, Annotations, contrasting colors, visual hierarchy 

```python
# Sample data
tips = sns.load_dataset('tips')

# Highlighting a specific data point
plt.figure(figsize=(10, 6))
sns.barplot(x='day', y='total_bill', data=tips, estimator=sum, palette='muted')

# Highlight Thursday
highlight = tips[tips['day'] == 'Thur']['total_bill'].sum()
plt.bar('Thur', highlight, color='red')

plt.title('Total Bill per Day')
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill (Sum)')
plt.show()
```

```python
# Illustrating annotations 
```

```python
# Generate sample data
np.random.seed(0)
x = np.random.rand(10)
y = np.random.rand(10)
sizes = 1000 * np.random.rand(10)  # Bubble sizes

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(x, y, s=sizes, alpha=0.5, c='blue')
plt.title('Scatter Plot with Annotations')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Adding annotations
for i, txt in enumerate(range(10)):
    plt.annotate(f'Point {txt}', (x[i], y[i]),
                 textcoords="offset points",
                 xytext=(0, 10),
                 ha='center',
                 fontsize=9,
                 color='red')

# Highlight a specific point
highlight_idx = 5
plt.scatter(x[highlight_idx],
            y[highlight_idx],
            s=sizes[highlight_idx],
            color='red')
plt.annotate('Highlighted Point', (x[highlight_idx], y[highlight_idx]),
             textcoords="offset points",
             xytext=(0, 15),
             ha='center',
             fontsize=12,
             color='red',
             weight='bold')

plt.grid(True)
plt.show()
```

## Telling a story

Intuition: Tell a compelling story, lead your viewers, create compelling narratives 
    
    
To achieve this: Context, flow, Emotions, Engagement 

```python
# Sample data
data = pd.DataFrame({
    'Year': [2017, 2018, 2019, 2020, 2021],
    'Sales': [350, 400, 450, 500, 600]
})

# Storytelling with a line plot
plt.figure(figsize=(10, 6))
plt.plot(data['Year'], data['Sales'], marker='o', color='b')
plt.title('Annual Sales Trend', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Sales (in thousands)', fontsize=14)

# Adding annotations to tell a story
for i, txt in enumerate(data['Sales']):
    plt.annotate(txt, (data['Year'][i], data['Sales'][i]),
                 textcoords="offset points",
                 xytext=(0, 10),
                 ha='center')

plt.grid(True)
plt.show()
```
