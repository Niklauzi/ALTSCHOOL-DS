```python
# experience level and employment type

experience_dictionary = {'EN':'Entry-level/Junior',
                         'MI':'Mid-level/Intermediate',
                         'SE':'Senior-level/Expert',
                         'EX':'Executive-level/Director'}
```

```python
df['experience_level'].str.replace('EN','Entry-level/Junior')
```

```python
df['experience_level'] = df['experience_level'].map(experience_dictionary)
```
