```python
import pandas as pd
```

```python
def add_product():
    # Prompt user for product name
    product_name = input("Enter product name: ")
    
    try:
        # Attempt to read the CSV file into a pandas DataFrame
        df = pd.read_csv('inentory_data.csv', header=None, names=['Product Name', 'Quantity', 'Price'])
        
        # Check if the product already exists
        existing_product = df[df['Product Name'].str.lower() == product_name.lower()]
        
        # If the product exists
        if not existing_product.empty:
            print(f"Product '{product_name}' already exists in the inventory.")
            update_option = input('Do you want to update the existing product? (y/n): ').lower()
            
            # If user chooses to update the existing product
            if update_option == 'y':
                # Get new quantity and price
                new_quantity = int(input('Enter new quantity: '))
                new_price = float(input('Enter new price per unit: '))
                
                # Update quantity and price for the existing product
                df.loc[df['Product Name'].str.lower() == product_name.lower(), 
                       ['Quantity', 'Price']] = [new_quantity, new_price]
                
                # Save the updated DataFrame back to the CSV file
                df.to_csv('inentory_data.csv', index=False)
                print(f'Product: {product_name} updated successfully.')
                print(df)
                return
            else:
                # If user chooses not to update, exit the function
                print('No changes made.')
                return
    except FileNotFoundError:
        # If the file doesn't exist, create a new DataFrame with column names
        df = pd.DataFrame(columns=['Product Name', 'Quantity', 'Price'])
    
    # Prompt user for quantity and price
    quantity = int(input('Enter quantity: '))
    price = float(input('Enter price per unit: '))
    
    # Add new product to the DataFrame
    new_product = pd.DataFrame({'Product Name': [product_name], 'Quantity': [quantity], 'Price': [price]})
    df = pd.concat([df, new_product], ignore_index=True)
    
    # Print the updated DataFrame
    print(df)
    
    # Save the updated DataFrame back to the CSV file
    df.to_csv('inentory_data.csv', index=False)
    
    print(f'Product: {product_name} added successfully!\n')
    
    # Print the DataFrame after adding the new product
    print(df)
    
#View Inventory
def view_inventory():
    try:
        df =pd.read_csv('inentory_data.csv') #header=None, names=['Product Name', 'Quantity', 'Price'])
        if df.empty:
            print('No products in the inventory')
        else:
            print('\n')
            print('Inventory:')
            print(df)
    except FileNotFOundError:
        print('No products in the inventory')

#search for a product
def search_product():
    search_name = input('Enter the name of the product to search: ')
    try:
        df =pd.read_csv('inentory_data.csv', header=None, names=['Product Name', 'Quantity', 'Product'])
       
        result = df[df['Product Name'].str.lower() ==search_name.lower()]
    
        if result.empty:
            
            print(f"No product with the name: '{search_name}'")
        else:
            
            print('product found')
            print(result)
    except FileNotFOundError:
        print('No products in the inventory')
```

```python
def rms():
    while True:  
        print("\nInventory Management System")
        print("1. Add Product")
        print("2. View Product")
        print("3. Search Product")
        print("4. Delete Product")
        print("5. Exit")
    
        choice = input("Enter your choice from (1-4): ")
        if choice == '1':
            add_product()
        elif choice == '2':
            view_inventory()
        elif choice == '3':
            search_product()
        elif choice == '4':
            delete_product()
        elif choice == '5':
            break
        else:
            print('Invalid choice. please enter a number between and 4.')
    
```

```python
def delete_product():
    product_name= input('Enter product name')
    try:
        # Attempt to read the CSV file into a pandas DataFrame
        df = pd.read_csv('inentory_data.csv', header=None, names=['Product Name', 'Quantity', 'Price'])
        
        # Check if the product already exists
        result = df[df['Product Name'].str.lower() == product_name.lower()]
        
        if result.empty:
            print(f"No product with the name '{product_name}'")
        else:
            df = df[df['Product Name'].str.lower() != product_name.lower()]
            df.to_csv('inentory_data.csv', index=False)
            
            # notify the user that the change has been made
            print(f"Product: '{product_name}' deleted successfully ")
    except FileNotFoundError:
        print('No products in the inventory')
            
            
            
```
